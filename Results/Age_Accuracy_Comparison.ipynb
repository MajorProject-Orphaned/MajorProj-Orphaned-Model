SAMPLE TESTING

```python
import os
folders = os.listdir(folderpath)
folders.sort()
for j in range(0,2):
  folder=f"{folders[j]}"
  dict_1={}
  files = os.listdir(f"{folderpath}/{folder}")
  files.sort()
  image_1 = f"{folderpath}/{folder}/{files[2]}"
  age_1=files[2][4:-4]
  age_1=age_1[0:2]
  print(f"Person Number {folder}")
  for i in range(1,len(files)):
    result=DeepFace.verify(image_1,f"{folderpath}/{folder}/{files[i]}",model_name="VGG-Face",detector_backend="mtcnn")
    current_age=files[i][4:-4]
    current_age=current_age[0:2]
    age_gap=abs(int(current_age)-int(age_1))
    if result['verified']==True:
      dict_1[f'Age_Gap {age_gap}']=1
    else:
      dict_1[f'Age_Gap {age_gap}']=0
  my_dict[f"Person {j}"]=dict_1

print(my_dict)
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(my_dict, orient='index')

df
```
